```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Load data
district = pd.read_csv("dataset/district.csv", sep=";")
```

```python
# 3D plot
fig = plt.figure(figsize=(8,8))

ax = fig.add_subplot(111, projection='3d')

ax.set_title("Number of municipalities with number of inhabitants in certain ranges", fontdict={"fontsize":14, "fontweight": "bold"})

ax.set_xlabel("< 499 inhabitants")

ax.set_ylabel("500-1999 inhabitants")

ax.set_zlabel("2000-9999 inhabitants")
ax.scatter(district["no. of municipalities with inhabitants < 499 "], district["no. of municipalities with inhabitants 500-1999"], district["no. of municipalities with inhabitants 2000-9999 "])
plt.savefig("new_stats/3d_district_municipalities.png")
```
