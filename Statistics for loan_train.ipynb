```python
# Imports
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Load data
loan_clean = pd.read_csv("new_script/loans_train.csv")
```

```python
# Distribution of owner age at loan
plt.title("Distribution of owner age at loan", fontdict={"fontsize":14, "fontweight": "bold"})
sns.histplot(data=loan_clean, x="owner_age_at_loan")
```

```python
# Distribution of owner age at account creation
plt.title("Distribution of owner age at account creation", fontdict={"fontsize":14, "fontweight": "bold"})
sns.histplot(data=loan_clean, x="owner_age_at_account_creation")
```

```python
# Booleans
cols = ["has_disponent", "has_junior_card", "has_classic_card", "has_gold_card"]
bools = loan_clean[cols]
num_entries = len(bools)
percentages = []
for i in bools:
    num_true = len(bools.loc[bools[i] == 1])
    percentages.append(num_true / num_entries * 100)
for c, i in enumerate(percentages):
    y = np.array([i, 100 - i])
    
    main_elem_of_label = str(cols[c]).split('_')[1]
    if c > 0:
        main_elem_of_label += " card"
    
    first_label = f"With {main_elem_of_label}"
    second_label = f"Without {main_elem_of_label}"
    labels = [first_label, second_label]
    
    plt.title(f"Distribution of account owners with {main_elem_of_label}", fontdict={"fontsize":14, "fontweight": "bold"})
    plt.pie(y, labels=labels, autopct='%.1f%%')
    plt.show()
```
